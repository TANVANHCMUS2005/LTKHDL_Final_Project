Kiểu dữ liệu không phù hợp: `product_price`,  `product_num_ratings` 

Chuyển `product_price` từ str sang float

Chuyển `product_num_ratings`  từ float sang int


```python
def process_price_to_usd(row):
    price_str = str(row['product_price'])
    country = row['country']
    
    # 1. LÀM SẠCH (CLEANING)
    # Giữ lại số và dấu câu
    clean_str = ''.join(c for c in price_str if c.isdigit() or c in [',', '.'])
    if not clean_str: return np.nan

    try:
        # Xử lý định dạng số theo vùng miền
        if country in ['DE', 'FR', 'IT', 'ES']: # Châu Âu (dùng phẩy làm thập phân)
            clean_str = clean_str.replace('.', '').replace(',', '.')
        else: # Mỹ, Á, Úc (dùng chấm làm thập phân)
            clean_str = clean_str.replace(',', '')
        
        raw_price = float(clean_str)
    except ValueError:
        return np.nan

    # 2. QUY ĐỔI TỶ GIÁ (CONVERSION) -> USD
    rates = {
        'US': 1.0,    # Gốc
        'CA': 0.74,   # Canada
        'AU': 0.66,   # Úc
        'MX': 0.058,  # Mexico
        'IN': 0.012,  # Ấn Độ
        'JP': 0.0067, # Nhật
        'DE': 1.08, 'FR': 1.08, 'IT': 1.08, 'ES': 1.08 # Euro Zone
    }
    
    exchange_rate = rates.get(country, 1.0) # Mặc định là 1 nếu không tìm thấy
    return round(raw_price * exchange_rate, 2) # Làm tròn 2 số thập phân

df['product_price'] = df.apply(process_price_to_usd, axis=1)
print('product_price:' ,df['product_price'].dtypes)
```

### 2.2. Numerical Columns Analysis 

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns
numerical_columns
```

```python
def analyze_numerical_column(df, col_name):
    print(f"{'='*40}")
    print(f" PHÂN TÍCH CỘT: {col_name}")
    print(f"{'='*40}")
    
    # Lấy dữ liệu (loại bỏ NaN để tính toán thống kê)
    data = df[col_name].dropna()
    
    # --- 1. DISTRIBUTION & CENTRAL TENDENCY ---
    mean = data.mean()
    median = data.median()
    std = data.std()
    skewness = data.skew()
    
    # Xác định hình dáng phân phối dựa trên độ lệch (Skewness)
    if skewness > 1: shape = "Lệch phải (Right Skewed)"
    elif skewness < -1: shape = "Lệch trái (Left Skewed)"
    else: shape = "Khá đối xứng (Normal-like)"
    
    print(f"1. Phân phối & Xu hướng tập trung:")
    print(f"   - Mean (TB): {mean:,.2f}")
    print(f"   - Median (Trung vị): {median:,.2f}")
    print(f"   - Std Dev (Độ lệch chuẩn): {std:,.2f}")
    print(f"   - Skewness: {skewness:.2f} -> Hình dạng: {shape}")

    # --- 2. RANGE & OUTLIERS (Phương pháp IQR) ---
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Đếm số lượng Outliers
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    print(f"2. Phạm vi & Giá trị ngoại lai (Outliers):")
    print(f"   - Min: {data.min():,.2f}")
    print(f"   - Max: {data.max():,.2f}")
    print(f"   - Số lượng Outliers: {len(outliers)}")


    # --- 3. DATA QUALITY ---
    missing_count = df[col_name].isnull().sum()
    missing_pct = (missing_count / len(df)) * 100
    
    # Kiểm tra giá trị vô lý (Impossible values)
    # Ví dụ: Giá < 0, hoặc Rating < 0
    negative_values = (data < 0).sum()
    zeros = (data == 0).sum()
    
    print(f"3. Chất lượng dữ liệu:")
    print(f"   - Giá trị thiếu: {missing_count} dòng ({missing_pct:.2f}%)")
    print(f"   - Giá trị âm (không hợp lệ): {negative_values}")
    print(f"   - Placeholder 0: {zeros}")


    # --- 4. VISUALIZATION ---
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # Biểu đồ Histogram (Phân phối)
    sns.histplot(data, kde=True, ax=axes[0], color='skyblue')
    axes[0].set_title(f'Histogram: {col_name}')
    
    # Biểu đồ Boxplot (Outliers)
    sns.boxplot(x=data, ax=axes[1], color='lightgreen')
    axes[1].set_title(f'Boxplot: {col_name}')
    
    plt.tight_layout()
    plt.show()
```

### 2.3. Categorical Columns Analysis 

```python
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
def analyze_categorical_column(df, col_name):
    print(f"\n{'='*50}")
    print(f" PHÂN TÍCH CỘT: {col_name}")
    print(f"{'='*50}")
    
    # --- A. VALUE DISTRIBUTION (Phân phối giá trị) ---
    # Đếm số lượng giá trị duy nhất
    unique_count = df[col_name].nunique()
    # Đếm tần suất xuất hiện của từng giá trị
    value_counts = df[col_name].value_counts()
    # Lấy Top 10 giá trị phổ biến nhất
    top_10 = value_counts.head(10)
    
    print("1. Phân Phối Giá Trị:")
    print(f"   - Số lượng giá trị duy nhất (Unique): {unique_count}")
    print(f"   - Giá trị phổ biến nhất (Top 1): '{top_10.index[0]}' ({top_10.iloc[0]} lần)")
    
    # Kiểm tra độ cân bằng (Balance Check)
    # Nếu giá trị xuất hiện nhiều nhất chiếm tỷ lệ quá lớn -> Imbalanced
    max_pct = value_counts.iloc[0] / len(df)
    if max_pct > 0.8:
        balance_status = "Mất cân bằng nghiêm trọng (Highly Imbalanced)"
    elif max_pct > 0.5:
        balance_status = "Mất cân bằng vừa phải"
    else:
        balance_status = "Khá cân bằng (hoặc phân tán đều)"
        
    if col_name == 'country': # Trường hợp đặc biệt của bộ dữ liệu này
        balance_status = "Cân bằng hoàn hảo (Do lấy mẫu Top 100/nước)"
        
    print(f"   - Đánh giá phân phối: {balance_status}")

    # --- B. DATA QUALITY (Chất lượng dữ liệu) ---
    # Tính giá trị thiếu
    missing = df[col_name].isnull().sum()
    missing_pct = (missing / len(df)) * 100
    
    # Kiểm tra tính nhất quán (Inconsistencies)
    # So sánh số lượng unique gốc với số lượng unique khi chuyển hết về chữ thường
    # Nếu khác nhau -> Có các giá trị giống nhau nhưng viết hoa/thường khác nhau (VD: "Male" vs "male")
    unique_lower = df[col_name].dropna().astype(str).str.lower().nunique()
    inconsistency_check = "Phát hiện không nhất quán (Case sensitivity)" if unique_lower < unique_count else "Không phát hiện lỗi viết hoa/thường"
    
    # Kiểm tra danh mục hiếm (Rare categories)
    # Đếm số lượng giá trị chỉ xuất hiện đúng 1 lần
    rare_count = (value_counts == 1).sum()
    
    print("2. Chất Lượng Dữ Liệu:")
    print(f"   - Giá trị thiếu (Missing): {missing} ({missing_pct:.2f}%)")
    print(f"   - Tính nhất quán: {inconsistency_check}")
    print(f"   - Số lượng danh mục hiếm (xuất hiện 1 lần): {rare_count}")
    if rare_count > 0:
        print("     -> Cần lưu ý: Quá nhiều danh mục hiếm có thể gây khó khăn cho mô hình học máy.")
        
    plt.figure(figsize=(12, 6))
    
    if col_name == 'country':
        # Vẽ biểu đồ đếm cho Quốc gia
        # Thêm hue=col_name và legend=False
        sns.countplot(y=col_name, data=df, order=value_counts.index, hue=col_name, legend=False, palette='viridis')
        plt.title(f'Phân bố sản phẩm theo {col_name}')
        plt.xlabel('Số lượng')
    elif col_name == 'product_title':
        # Với tên sản phẩm, chỉ vẽ Top 10 vì quá nhiều giá trị
        sns.barplot(x=top_10.values, y=top_10.index, hue=top_10.index, legend=False, palette='magma')
        plt.title(f'Top 10 Tên Sản Phẩm phổ biến nhất')
        plt.xlabel('Tần suất')
        
    plt.tight_layout()
    plt.show()
```

```python
developed_markets = ['US', 'DE', 'FR', 'IT', 'ES', 'JP', 'CA', 'AU', 'UK'] 
emerging_markets = ['IN', 'MX', 'BR'] 

def get_market_type(country_code):
    if country_code in developed_markets:
        return 'Developed'
    elif country_code in emerging_markets:
        return 'Emerging'
    else:
        return 'Other'
df_q1 = df_clean.copy()

df_q1['Market_Type'] = df_q1['country'].apply(get_market_type)

df_q1 = df_q1[df_q1['Market_Type'] != 'Other']

print("Số lượng mẫu theo nhóm thị trường:")
print(df_q1['Market_Type'].value_counts())

from scipy.stats import mannwhitneyu

sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Market_Type', y='product_price', data=df_q1, palette="Set2", showfliers=False)
plt.title('So sánh Mức Giá: Developed vs. Emerging', fontsize=14)
plt.ylabel('Giá sản phẩm (USD)', fontsize=12)
plt.xlabel('Loại thị trường', fontsize=12)

plt.subplot(1, 2, 2)
sns.boxplot(x='Market_Type', y='product_star_rating', data=df_q1, palette="Set2")
plt.title('So sánh Đánh giá (Rating): Developed vs. Emerging', fontsize=14)
plt.ylabel('Điểm đánh giá (1-5 sao)', fontsize=12)
plt.xlabel('Loại thị trường', fontsize=12)

plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("KẾT QUẢ KIỂM ĐỊNH THỐNG KÊ (MANN-WHITNEY U TEST)")
print("="*50)


group_dev = df_q1[df_q1['Market_Type'] == 'Developed']
group_eme = df_q1[df_q1['Market_Type'] == 'Emerging']


stat_price, p_price = mannwhitneyu(group_dev['product_price'], group_eme['product_price'], alternative='two-sided')

print(f"\n1. So sánh về GIÁ (Price):")
print(f"   - Statistic: {stat_price:.2f}")
print(f"   - P-value: {p_price:.4e}") 
if p_price < 0.05:
    print("   => KẾT LUẬN: Có sự khác biệt CÓ Ý NGHĨA THỐNG KÊ về giá.")
    mean_dev = group_dev['product_price'].median()
    mean_eme = group_eme['product_price'].median()
    print(f"      (Trung vị Giá Developed: ${mean_dev} vs. Emerging: ${mean_eme})")
else:
    print("   => KẾT LUẬN: KHÔNG có sự khác biệt về giá.")


stat_rating, p_rating = mannwhitneyu(group_dev['product_star_rating'], group_eme['product_star_rating'], alternative='two-sided')

print(f"\n2. So sánh về ĐÁNH GIÁ (Rating):")
print(f"   - Statistic: {stat_rating:.2f}")
print(f"   - P-value: {p_rating:.4f}")
if p_rating < 0.05:
    print("   => KẾT LUẬN: Có sự khác biệt CÓ Ý NGHĨA THỐNG KÊ về rating.")
    print("      (Người dùng ở một thị trường khắt khe hơn thị trường kia)")
else:
    print("   => KẾT LUẬN: KHÔNG có sự khác biệt về rating giữa 2 nhóm.")
```

### Nhận xét Kết quả Kiểm định (Statistical Interpretation)

Để đánh giá sự khác biệt về giá giữa hai nhóm thị trường, chúng tôi thiết lập bài toán kiểm định giả thuyết như sau:

**1.2.1 Thiết lập Giả thuyết (Hypothesis Setup):**
* **Giả thuyết không ($H_0$ - Null Hypothesis):** Không có sự khác biệt về phân phối giá sản phẩm giữa thị trường Phát triển (*Developed*) và thị trường Mới nổi (*Emerging*).
    * *(Hàm ý: Mọi sự chênh lệch giá quan sát được chỉ là do ngẫu nhiên)*.
* **Giả thuyết đối ($H_1$ - Alternative Hypothesis):** Có sự khác biệt có ý nghĩa thống kê về phân phối giá sản phẩm giữa hai nhóm thị trường này.
    * *(Hàm ý: Có yếu tố hệ thống tác động khiến giá ở hai thị trường khác nhau)*.

**1.2.2 Phân tích P-value:**
* Kết quả kiểm định Mann-Whitney U cho ra giá trị **P-value = $2.6246 \times 10^{-6}$** (tương đương $0.0000026$).
* So sánh với mức ý nghĩa tiêu chuẩn $\alpha = 0.05$, ta thấy: $P\text{-value} \ll 0.05$.

**1.2.3. Kết luận Kiểm định:**
* Do $P\text{-value} < 0.05$, chúng tôi có đủ cơ sở thống kê vững chắc để **bác bỏ giả thuyết không ($H_0$)** và **chấp nhận giả thuyết đối ($H_1$)**.
* **Kết luận:** Có sự khác biệt có ý nghĩa thống kê về mức giá trung bình giữa nhóm thị trường Phát triển và Mới nổi.

**1.2.4. Ý nghĩa Thực tiễn:**
* Kết quả này xác nhận rằng các công ty phần mềm đang áp dụng chiến lược định giá phân biệt (Price Discrimination): Thị trường phát triển (Mỹ, Châu Âu) có mức giá niêm yết cao hơn đáng kể, trong khi thị trường mới nổi được hưởng mức giá thấp hơn để phù hợp với khả năng chi trả.

```python
df_q2 = df_clean.copy()
df_q2['Rank'] = df_q2.groupby('country').cumcount() + 1


def categorize_rank(rank):
    if rank <= 10: return 'Top 10'
    elif rank <= 50: return 'Top 50'
    else: return 'Top 50-100'

df_q2['Rank_Group'] = df_q2['Rank'].apply(categorize_rank)

corr_star = df_q2['product_star_rating'].corr(df_q2['Rank'], method='spearman')
corr_num = df_q2['product_num_ratings'].corr(df_q2['Rank'], method='spearman')

print("="*50)
print("KẾT QUẢ TƯƠNG QUAN SPEARMAN (Rank càng nhỏ = Càng bán chạy)")
print("="*50)
print(f"1. Rating vs. Rank: {corr_star:.3f}")
print("   (Ý nghĩa: Điểm sao cao có giúp Rank giảm/tốt hơn không?)")
print(f"2. Num Ratings vs. Rank: {corr_num:.3f}")
print("   (Ý nghĩa: Nhiều review có giúp Rank giảm/tốt hơn không?)")

if abs(corr_num) > abs(corr_star):
    print("\n=> KẾT LUẬN SƠ BỘ: 'Hiệu ứng đám đông' (Num Ratings) ảnh hưởng mạnh hơn Chất lượng.")
else:
    print("\n=> KẾT LUẬN SƠ BỘ: 'Chất lượng' (Star Rating) ảnh hưởng mạnh hơn.")

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.regplot(x='product_star_rating', y='Rank', data=df_q2, 
            scatter_kws={'alpha':0.3, 'color':'blue'}, line_kws={'color':'red'})
plt.title(f'Tác động của Chất lượng (Rating)\nCorr: {corr_star:.3f}', fontsize=12)
plt.ylabel('Thứ hạng Bestseller (Càng thấp càng tốt)')
plt.xlabel('Điểm đánh giá (Sao)')
plt.gca().invert_yaxis() 

plt.subplot(1, 2, 2)
sns.regplot(x='product_num_ratings', y='Rank', data=df_q2, 
            scatter_kws={'alpha':0.3, 'color':'green'}, line_kws={'color':'red'})
plt.title(f'Tác động của Đám đông (Num Ratings)\nCorr: {corr_num:.3f}', fontsize=12)
plt.xlabel('Số lượng đánh giá (Log Scale)')
plt.ylabel('')
plt.xscale('log') 
plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()
```

    * Dữ liệu phân tán hỗn loạn, không tạo thành quy luật rõ ràng.
    * Đường hồi quy (Regression Line) gần như nằm ngang.
    * **Thực tế quan sát:** Rất nhiều sản phẩm đạt điểm tuyệt đối 5.0 sao nhưng lại nằm ở vị trí thấp (Rank > 80). Ngược lại, nhiều sản phẩm trong Top 10 chỉ có điểm trung bình khá (khoảng 4.3 - 4.5).

* **Biểu đồ 2: Rank vs. Num Ratings (Bên phải)**
    * Có xu hướng tập trung rõ rệt hơn: Các sản phẩm nằm ở vùng **Top Rank (1-10)** thường sở hữu lượng đánh giá rất lớn (hàng nghìn review).
    * Khi số lượng đánh giá giảm xuống thấp, sản phẩm có xu hướng trượt dài trên bảng xếp hạng (Rank tăng dần).
    * Đường hồi quy dốc xuống, xác nhận mối quan hệ: **Tăng Volume review $\rightarrow$ Cải thiện Ranking.**

#### c. Kết luận & Ý nghĩa Thực tiễn (Conclusion & Business Implications)

**1. Kết luận:**
Dữ liệu khẳng định rằng trong cuộc đua Bestseller, **"Hiệu ứng đám đông" (Social Proof) chiến thắng "Chất lượng thuần túy"**. Người dùng có xu hướng tin tưởng và mua các sản phẩm "đã được kiểm chứng" bởi nhiều người khác, hơn là mạo hiểm mua một sản phẩm điểm cao nhưng ít người biết đến.

**2. Chiến lược cho Doanh nghiệp:**
* **Ưu tiên số lượng:** Đối với sản phẩm mới, mục tiêu sống còn là tăng nhanh **số lượng đánh giá** (Volume) để vượt qua ngưỡng tin cậy của khách hàng và thuật toán đề xuất.
* **Đừng ám ảnh sự hoàn hảo:** Một sản phẩm 4.4 sao với 10.000 đánh giá sẽ mang lại doanh thu tốt hơn nhiều so với sản phẩm 5.0 sao nhưng chỉ có 50 đánh giá.
* **Hành động:** Thay vì chi quá nhiều tiền R&D để nâng điểm số từ 4.5 lên 4.8, hãy dùng ngân sách đó cho Marketing/Seeding để khuyến khích khách hàng để lại review.
